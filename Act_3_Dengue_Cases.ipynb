<h2>Q11: Dengue Cases per Year</h2>

```python
cases_per_year = data.groupby('Year')['Dengue_Cases'].sum()
cases_per_year.plot(kind='bar', color='cyan', edgecolor = 'black')

plt.title('Total Dengue Cases per Year')
plt.xlabel('Year')
plt.ylabel('Total Dengue Cases')
plt.show()
```

<h3>Insights: This bar chart shows how dengue cases vary each year. You can easily see which year had the highest outbreak and if cases are increasing or decreasing over time.</h3><hr>

<h3>Q12: Total Dengue Deaths by Region</h3>

```python
deaths_by_region = data.groupby('Region')['Dengue_Deaths'].sum()
deaths_by_region.plot(kind='bar', color='cyan', edgecolor = 'black')

plt.title('Total Dengue Deaths by Region')
plt.xlabel('Region')
plt.ylabel('Total Deaths')
plt.show()
```

<h3>Insights: This visualization shows which regions recorded the most dengue deaths, helping identify areas with the highest mortality due to dengue.</h3><hr>

<h3>Q13: Average Dengue Cases per Month</h3>

```python
avg_monthly_cases = data.groupby('Month')['Dengue_Cases'].mean()
avg_monthly_cases.plot(kind='bar', color='cyan', edgecolor ="black")

plt.title('Average Dengue Cases per Month')
plt.xlabel('Month')
plt.ylabel('Average Cases')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

<h3>Insights:></h3><hr>

<h2>Q14: Dengue Cases vs. Deaths (Scatter Plot)</h2>

```python
sns.scatterplot(data=data, x='Dengue_Cases', y='Dengue_Deaths', hue='Region')
plt.title('Relationship Between Dengue Cases and Deaths')
plt.xlabel('Dengue Cases')
plt.ylabel('Dengue Deaths')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

<h3>Insights: This scatter plot shows whether regions with more dengue cases also have higher deaths. A rising pattern suggests that more cases often lead to more fatalities.</h3><hr>

<h3>Q15: Percentage of Total Dengue Cases by Region (Pie Chart)</h3>

```python
yearly_cases = data.groupby('Year')['Dengue_Cases'].sum()

yearly_cases.plot(kind='line', marker='o', color='blue', linewidth=2,)
plt.title('Total Dengue Cases per Year')
plt.xlabel('Year')
plt.ylabel('Total Dengue Cases')
plt.grid(True, linestyle='--', alpha=0.6)
plt.gcf().set_size_inches(8,5)
plt.show()
```

<h3>Insights: This line graph shows how dengue cases change year by year.
The grid lines make it easier to compare values, while the orange line highlights trends in outbreaks more clearly.</h3><hr>
